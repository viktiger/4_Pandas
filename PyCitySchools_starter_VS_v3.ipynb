```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="outer", on=["school_name", "school_name"])

school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools
Total_schools_df = school_data_complete["school_name"].unique()
Total_schools_count_df = len(Total_schools_df)
# Total_schools_count_df

# Calculate the total number of students
Total_students_df = school_data_complete["student_name"].count()
Total_students_df = '{:,.0f}'.format(Total_students_df)
# Total_students_df

# Calculate the total budget
Total_budget_df = school_data_complete["budget"].unique()
Total_budget_sum_df = Total_budget_df.sum()
Total_budget_sum_df = '${:,.0f}'.format(Total_budget_sum_df)
# Total_budget_sum_df

# Calculate the average math score
Average_math_df = school_data_complete["math_score"].mean()
Average_math_df = '{:,.6f}'.format(Average_math_df)
# Average_math_df

# Calculate the average reading score
Average_reading_df = school_data_complete["reading_score"].mean()
Average_reading_df = '{:,.5f}'.format(Average_reading_df)
# Average_reading_df

# Calculate the percentage of students with a passing math score (70 or greater)
Greaterthan70_math_df = school_data_complete.loc[school_data_complete["math_score"] >=70,:]
Greaterthan70_math_average_df = Greaterthan70_math_df["math_score"].count() / school_data_complete["student_name"].count()
Greaterthan70_math_average_df = '{:,.6%}'.format(Greaterthan70_math_average_df)
# Greaterthan70_math_average_df

# Calculate the percentage of students with a passing math score (70 or greater)
Greaterthan70_reading_df = school_data_complete.loc[school_data_complete["reading_score"] >=70,:]
Greaterthan70_reading_average_df = Greaterthan70_reading_df["reading_score"].count() / school_data_complete["student_name"].count()
Greaterthan70_reading_average_df = '{:,.6%}'.format(Greaterthan70_reading_average_df)
# Greaterthan70_reading_average_df

# Calculate the percentage of students who passed math and reading (% Overall Passing)
Greaterthan70_both_df = school_data_complete.loc[(school_data_complete["math_score"] >=70)&(school_data_complete["reading_score"] >=70)]
# Greaterthan70_both_df
Greaterthan70_both_overall_df = Greaterthan70_both_df["student_name"].count() / school_data_complete["student_name"].count()
Greaterthan70_both_overall_df = '{:,.6%}'.format(Greaterthan70_both_overall_df)
# Greaterthan70_both_overall_df

# Create a dataframe to hold the above results
Summary_df = pd.DataFrame({"Total Schools":[Total_schools_count_df],
                           "Total Students":[Total_students_df],
                           "Total Budget":[Total_budget_sum_df],
                           "Average Math Score":[Average_math_df],
                           "Average Reading Score":[Average_reading_df],
                           "Passing Math Score":[Greaterthan70_math_average_df],
                           "Passing Reading Score":[Greaterthan70_reading_average_df],
                           "% Overall Passing":[Greaterthan70_both_overall_df]})
Summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Create an overview table that summarizes key metrics about each school, including: 
# School Name, Total Students, Total School Budget, Average Math Score, Average Reading Score
Summary_List = school_data_complete.groupby("school_name").mean()
del Summary_List["Student ID"]
del Summary_List["School ID"]
Summary_List = Summary_List.rename(columns={"school_name":"School Name", 
                                            "size":"Total Students",
                                            "type":"School Type",
                                            "budget":"Total School Budget", 
                                            "math_score": "Average Math Score", 
                                            "reading_score":"Average Reading Score"})
Summary_List.head()

# Create an overview table that summarizes key metrics about each school, including: 
# Per Student Budget
Per_Student_Budget = Summary_List["Total School Budget"]/Summary_List["Total Students"]
Summary_List["Per Student Budget"] = Per_Student_Budget
# Summary_List.head()

# Create an overview table that summarizes key metrics about each school, including: 
# % Passing Math
Greaterthan70_math_df = school_data_complete.loc[school_data_complete["math_score"] >=70,:]
# Greaterthan70_math_df
Greaterthan70_math_count = Greaterthan70_math_df.groupby("school_name").count()
# Greaterthan70_math_count
Greaterthan70_math_percent = Greaterthan70_math_count["math_score"] / Summary_List["Total Students"]*100
# Greaterthan70_math_percent

Summary_List["% Passing Math"] = Greaterthan70_math_percent
# Summary_List.head()

# Create an overview table that summarizes key metrics about each school, including: 
# % Passing Reading
Greaterthan70_reading_df = school_data_complete.loc[school_data_complete["reading_score"] >=70,:]
Greaterthan70_reading_count = Greaterthan70_reading_df.groupby("school_name").count()
Greaterthan70_reading_percent = Greaterthan70_reading_count["reading_score"] / Summary_List["Total Students"]*100

Summary_List["% Passing Reading"] = Greaterthan70_reading_percent
# Summary_List.head()

# Create an overview table that summarizes key metrics about each school, including: 
# % Overall Passing
Greaterthan70_both_df = school_data_complete.loc[(school_data_complete["math_score"] >=70)&(school_data_complete["reading_score"] >=70)]
# Greaterthan70_both_df
Greaterthan70_both_count = Greaterthan70_both_df.groupby("school_name").count()
# Greaterthan70_both_count
Student_grouped_list = student_data.groupby("school_name").count()
# Student_grouped_list
Greaterthan70_both_percent = Greaterthan70_both_count / Student_grouped_list *100
# Greaterthan70_both_percent["student_name"]

Summary_List["% Passing Overall"] = Greaterthan70_both_percent["student_name"]
# Summary_List.head()

# Create an overview table that summarizes key metrics about each school, including: 
# School Type
School_Type = school_data_complete.loc[:,["school_name" , "type"]]
Summary_List_Type = pd.merge(Summary_List, School_Type, on="school_name", how="inner")
Summary_List_Type.drop_duplicates(inplace=True)
Summary_List_Type.reset_index(drop=True, inplace=True)

Summary_List_Type = Summary_List_Type.rename(columns={"type":"School Type"})
# Summary_List_Type.head()

# Organise & reorder columns
Summary_List_Organise = Summary_List_Type[["school_name","School Type","Total Students","Total School Budget","Per Student Budget","Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Passing Overall"]]
Summary_List_Organise
```

```python
# Format output
Summary_List_Organise.style.format({"Total Students": "{:.0f}", 
                                    "Total School Budget": "${:,.2f}", 
                                    "Per Student Budget": "${:.2f}", 
                                    "Average Math Score": "{:.6f}", 
                                    "Average Reading Score": "{:.6f}", 
                                    "% Passing Math": "{:.6f}", 
                                    "% Passing Reading": "{:.6f}", 
                                    "% Overall Passing": "{:.6f}"})
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Create a table that lists the average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school
# Group each series by school
Grade_nine_math = student_data.loc[student_data["grade"] == "9th"].groupby("school_name")
Grade_nine_math_mean = Grade_nine_math["math_score"].mean()

Grade_ten_math = student_data.loc[student_data["grade"] == "10th"].groupby("school_name")
Grade_ten_math_mean = Grade_ten_math["math_score"].mean()

Grade_eleven_math = student_data.loc[student_data["grade"] == "11th"].groupby("school_name")
Grade_eleven_math_mean = Grade_eleven_math["math_score"].mean()

Grade_twelve_math = student_data.loc[student_data["grade"] == "12th"].groupby("school_name")
Grade_twelve_math_mean = Grade_twelve_math["math_score"].mean()

# Combine the series into a dataframe
Math_scores_by_grade = pd.DataFrame({"9th": Grade_nine_math_mean, "10th": Grade_ten_math_mean, "11th": Grade_eleven_math_mean, "12th": Grade_twelve_math_mean})
# Math_scores_by_grade.index.name = "School"

# Formatting
Math_scores_by_grade.style.format({"9th": "{:.3f}", 
                                   "10th": "{:.3f}", 
                                   "11th": "{:.3f}", 
                                   "12th": "{:.3f}"})
```
